```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
R0 = 2200

Rp = 200e3 
def Rpara(R1,R2):
    return (R1*R2)/(R1+R2)
Rinc = np.arange(0,int(Rp/2),1)
Rc1 =  Rpara(R0, Rp/2 + Rinc)
Rc2 =  Rpara(R0, Rp/2 - Rinc)
print(Rc1)
plt.figure()
plt.plot(Rc1, label = 'Rc1')
plt.plot(Rc2, label = 'Rc2')
# plt.y('log')
plt.legend()
```

```python
def asym(R1,R2):
    print((R1 - R2)*100 / R1)
```
