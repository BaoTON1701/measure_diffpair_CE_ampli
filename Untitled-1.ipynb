```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
import re

def extract_noise_data(input_file, output_file):
    # Regex to match the Noise Contribution value (scientific notation)
    # and the value of M appearing later in the block
    noise_pattern = re.compile(r"rn\s+([\d\.e-]+)\s+100\.00")
    m_pattern = re.compile(r"with\s+M\s+=\s+([\d\.]+)")

    results = []
    
    with open(input_file, 'r') as f:
        content = f.read()
        
        # We split by the "Device" keyword to handle each block individually
        blocks = content.split("Device")
        
        for block in blocks:
            noise_match = noise_pattern.search(block)
            m_match = m_pattern.search(block)
            
            if noise_match and m_match:
                noise_val = noise_match.group(1)
                m_val = m_match.group(1)
                results.append(f"{m_val}\t{noise_val}")

    # Write the results to a new text file
    with open(output_file, 'w') as out:
        out.write("M\tnoise\n")  # Header
        out.write("\n".join(results))

    print(f"Extraction complete. Data saved to {output_file}")

# Usage
extract_noise_data('../simu/noise_summary_R6.txt', 'noise_data.txt')
```

```python
noise_data = pd.read_csv('noise_data.txt', sep='\s+',  engine='python')
noise_data.head()
```

```python
G = noise_gain.iloc[1].values[1:]
noise_data['G'] = G 
noise_data['CMRR'] = 4e-9 * noise_data['G'] / noise_data['noise']
```
